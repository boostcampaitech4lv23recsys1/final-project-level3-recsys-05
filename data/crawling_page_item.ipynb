```python
import requests
import pandas as pd
import time
import random

header = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}

pages = 7932
page_ids = []
for page in range(pages // 200 + 1) :
    base = f"https://ohou.se/projects.json?writer=self&page={page+1}&per=200&v=3"
    request = requests.get(base, headers=header, timeout=5)
    request_json = request.json()
    for item in request_json["projects"] :
        page_ids.append(item['id'])

len(page_ids)
```

```python
pages = pd.DataFrame(page_ids, columns=['page_id'])
pages.to_csv('page_ids.csv', index=False)
```

```python
page_category = []
product_ids = []

count = 0

for id in page_ids :
    page = f"https://ohou.se/projects/{id}/detail.json?v=4&"
    request = requests.get(page, headers=header, timeout=5)
    request_json = request.json()['project']
    page_category.append([id, request_json['title'], request_json['created_at'], request_json['residence'], request_json['area'], 
    request_json['agent'], request_json['expertise'], request_json['family_list'], request_json['style_list'], request_json['budget']])
    for product in request_json['bucketplace_document_supplements']['products'] :
        product_ids.append([id, product['id']])

    count += 1
    if count % 100 == 0 :
        time.sleep(random.randint(10, 20))

len(page_category), len(product_ids)
```

```python
page_category = pd.DataFrame(page_category, columns=['page_id', 'title', 'timestamp', 'residence', 'area', 'agent', 'expertise', 'family_list', 'style_list', 'budget'])
page_category.to_csv('pages.csv', index=False)
page_category
```

```python
product_ids = pd.DataFrame(product_ids, columns=['page_id', 'product_id'])
product_ids.to_csv('page_products.csv', index=False)
product_ids
```
