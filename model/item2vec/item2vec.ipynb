```python
data = pd.read_csv('../../fastapi_test/ohou-v4.inter', lineterminator='\n')
data
```

```python
data = data[['item_id:token', 'user_id:token', 'timestamp:float']]
data.columns = ['item_id', 'user_id', 'time']
data = data.sort_values(['user_id', 'time']).reset_index(drop=True)
data
```

```python
items = []
temp_items = []
pre = data.loc[0, 'user_id']
for _, (item, user) in data[['item_id', 'user_id']].iterrows() :
    if pre != user :
        items.append(temp_items)
        pre = user
        temp_items = []

    temp_items.append(item)
items.append(temp_items)

items
```

```python
n = 107013
for i in [50, 100] :
    for j in [7, 9, 11]:
        temp_model = Word2Vec(items, sg=1, vector_size=i, window=j)
        temp_item_vectors = temp_model.wv
        print(f"vector_size {i}, window {j} = {[(product_si[product_si['item_id']==id].title.values, sim) for id, sim in temp_item_vectors.most_similar(n)]}")
```

```python
model = Word2Vec(items, sg=1, vector_size=100, window=9)
model
```

```python
item_vectors = model.wv
item_vectors
```

```python
n = 107013
index = [i for i, _ in item_vectors.most_similar(n, topn=10)]
index.append(n)
product_si[product_si['item_id'].isin(index)]
```
